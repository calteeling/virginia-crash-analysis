```python
import pandas as pd
import folium
from folium.plugins import HeatMap
import plotly.express as ex
import os
```

```python
df = pd.read_csv('../data/cleaned_crash_data.csv')
df.head()
```

```python
df.info()
df.describe()
df[['Latitude', 'Longitude']].isna().sum()
```

```python
df = df.dropna(subset=['Latitude', 'Longitude'])
df = df[(df['Latitude'] != 0) & (df['Longitude'] != 0)]
```

```python
virginia_center = [37.5, -78.7]
crash_map = folium.Map(location=virginia_center, zoom_start=6, tiles='CartoDB positron')
```

```python
severity_weights = {
    'Property Damage Only': 1,
    'Injury': 3,
    'Serious Injury': 5,
    'Fatal': 7
}

df['SeverityWeight'] = df['SeverityLabel'].map(severity_weights)

from folium.plugins import HeatMap

heatmap_data = df[['Latitude', 'Longitude', 'SeverityWeight']].dropna().values

heatmap_map = folium.Map(
    location=[df['Latitude'].mean(), df['Longitude'].mean()],
    zoom_start=7,
    tiles='OpenStreetMap'
)

HeatMap(
    data=heatmap_data,
    radius=6,
    blur=8,
    min_opacity=0.5,
    gradient = {
    0.0: 'navy',
    0.2: 'blue',
    0.4: 'lime',
    0.6: 'yellow',
    0.8: 'orange',
    1.0: 'red'
}

).add_to(heatmap_map)

heatmap_map
```
